```python
# Run a grid experiment similar to run_w1_grid.sh and visualize results
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys

# Add parent directory to path for importing
sys.path.append('..')
from thesis.fuzzy.membership import compute_ndg_streaming
from thesis.exp.ndg_vs_kde import _kde_density, load_dataset_data

os.environ["THESIS_DATA"] = "/Users/nterlemes/personal/thesis/Data"
print(os.environ.get('THESIS_DATA'))
# Parameters
DATASETS = ['opportunity', 'pamap2']
NUM_FOLDS = 3
SIGMAS = [0.2, 0.4, 0.8]

# Create results directory if it doesn't exist
os.makedirs('results', exist_ok=True)
OUTPUT_FILE = 'results/rq1_grid_notebook.csv'

# Initialize results DataFrame with header
results_df = pd.DataFrame(columns=['dataset', 'fold', 'sigma', 'kl_div', 'chi2', 'wall_sec', 'peak_rss_mb'])

for dataset in DATASETS:
    for fold in range(NUM_FOLDS):
        for sigma in SIGMAS:
            print(f"▶ {dataset} fold={fold}/{NUM_FOLDS} σ={sigma}")
            
            # Instead of using subprocess, let's call the function directly
            import thesis.exp.ndg_vs_kde as ndg_vs_kde
            
            try:
                # Call run_once directly
                kl_div, chi2 = ndg_vs_kde.run_once(dataset, fold, sigma, NUM_FOLDS)
                wall_sec = 0.0  # We won't measure these when calling directly
                peak_rss_mb = 0.0
                
                # Add to results DataFrame
                new_row = pd.DataFrame([{
                    'dataset': dataset,
                    'fold': fold,
                    'sigma': sigma,
                    'kl_div': float(kl_div),
                    'chi2': float(chi2),
                    'wall_sec': float(wall_sec),
                    'peak_rss_mb': float(peak_rss_mb)
                }])
                
                results_df = pd.concat([results_df, new_row], ignore_index=True)
                
                print(f"  KL: {float(kl_div):.4f}, Chi²: {float(chi2):.4f}")
            except Exception as e:
                print(f"Error running experiment: {e}")
                
# Save results to CSV
results_df.to_csv(OUTPUT_FILE, index=False)
print(f"✅ Grid finished → {OUTPUT_FILE}")

# Visualize results
plt.figure(figsize=(15, 10))

# Create subplots for KL divergence and Chi-squared metrics
plt.subplot(2, 2, 1)
sns.boxplot(x='sigma', y='kl_div', hue='dataset', data=results_df)
plt.title('KL Divergence by Dataset and Sigma')
plt.xlabel('Sigma Value')
plt.ylabel('KL Divergence')

plt.subplot(2, 2, 2)
sns.boxplot(x='sigma', y='chi2', hue='dataset', data=results_df)
plt.title('Chi-squared by Dataset and Sigma')
plt.xlabel('Sigma Value')
plt.ylabel('Chi-squared')

# Plot by fold for KL divergence
plt.subplot(2, 2, 3)
sns.lineplot(x='sigma', y='kl_div', hue='dataset', style='fold', data=results_df, markers=True)
plt.title('KL Divergence Trends by Fold')
plt.xlabel('Sigma Value')
plt.ylabel('KL Divergence')

# Plot computational metrics
plt.subplot(2, 2, 4)
sns.scatterplot(x='wall_sec', y='peak_rss_mb', hue='dataset', size='sigma', data=results_df)
plt.title('Computational Performance')
plt.xlabel('Wall Time (seconds)')
plt.ylabel('Peak Memory Usage (MB)')

plt.tight_layout()
plt.savefig('results/ndg_vs_kde_comparison.png', dpi=300)
plt.show()

# Visualize actual density functions for specific examples
plt.figure(figsize=(15, 15))

# Plot density functions for different sigma values
for i, sigma in enumerate(SIGMAS):
    for j, dataset in enumerate(DATASETS):
        plt.subplot(len(SIGMAS), len(DATASETS), i*len(DATASETS) + j + 1)
        
        # Load data for the first fold
        fold = 0
        data, x = load_dataset_data(dataset, fold, NUM_FOLDS)
        
        # Compute densities
        ndg = compute_ndg_streaming(x, data, sigma)
        kde = _kde_density(x, data, sigma)
        
        # Ensure both are normalized to integrate to 1.0
        ndg_integral = np.trapz(ndg, x=x)
        kde_integral = np.trapz(kde, x=x)
        
        if abs(ndg_integral - 1.0) > 0.01:
            ndg = ndg / ndg_integral
        
        if abs(kde_integral - 1.0) > 0.01:
            kde = kde / kde_integral
        
        # Plot
        plt.plot(x, ndg, 'b-', label='NDG')
        plt.plot(x, kde, 'r-', label='KDE')
        plt.fill_between(x, 0, ndg, alpha=0.3, color='blue')
        plt.fill_between(x, 0, kde, alpha=0.3, color='red')
        
        # Add histogram of original data for reference
        plt.hist(data, bins=30, density=True, alpha=0.2, color='gray')
        
        # Calculate difference metrics
        kl_div = np.sum(ndg * np.log((ndg + 1e-12) / (kde + 1e-12)))
        chi2 = np.sum(((ndg - kde) ** 2) / (kde + 1e-12))
        
        plt.title(f'{dataset.capitalize()}, σ={sigma}\nKL={kl_div:.4f}, Chi²={chi2:.4f}')
        plt.xlabel('Value')
        plt.ylabel('Density')
        plt.legend()

plt.tight_layout()
plt.savefig('results/density_functions_comparison.png', dpi=300)
plt.show()
```
